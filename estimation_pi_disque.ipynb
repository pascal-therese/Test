Créer une fonction renvoyant les coordonnées d''un point aléatoire du carré délimité par les points de coordonnées (-1;-1) et (1;1)
Il n'y a plus qu'à faire une fonction tirant un certain nombre de points au hasard afin d'estimer l'aire du disque à partir de la fréquence des points tombant dans le disque
Attention la puissance se note **2

```python
from random import *
from math import *
import matplotlib.pyplot as plt

def point_aleatoire():
    "Renvoie les coordonnées d'un point aléatoire du carré centré sur O et de coté 2"
    return  (-1+2*random(),-1+2*random())                          

def est_dans_le_disque(xA,yA):
    "Renvoie True si le point entré par ses coordonnées x,y est dans le disque de centre O et de rayon 1"
    assert(-1<=xA<=1)
    assert(-1<=yA<=1)
    if xA**2+yA**2<=1:
        return(True)
    else : 
        return(False)


    
def compteur_de_points(n=100):
    """ cette fonction renvoie le nombre de points dans le disque
        cette fonction dessine les points dans le disque"""
    assert(isinstance(n,int))
    c=0 # intiliatilsation du compteur
    for i in range(n):
        xA,yA=point_aleatoire()
        if est_dans_le_disque(xA,yA)==True:
            c=c+1
            plt.scatter(xA,yA,s=130)

    return(c)


        
```

```python
def demo():
    n=int(input("Entrer le nombre de points : "))
    nb_points=compteur_de_points(n)
    va=nb_points/n*4
    print("une valeur approchée de pi est : ",va)
    print("le nombre de points dans le disque est : ",nb_points)
    plt.axis('equal')

    # plt.savefig('PI.png')
    plt.show()
    return()

```

```python
demo()


```

Deuxième idée : trouver l'aire d'un quart de disque de rayon 1 en l'approchant par des rectangles.



On peut aussi ajouter une à une les aires de petits rectangles.!


```python
from math import *
def aire_du_disque(n=1000):
    "renvoie l'aire du disque estimée par la somme des aires de n rectangles dans un quart de cercle"
    #On calcule d'abord l'aire du quart de disque
    h=1.0/n   #la largeur d'un rectangle
    x=0        #l'abscisse du premier point en bas à gauche du rectangle
    A=0        #Somme des aires des rectangles 
    for k in range(n):
        y=sqrt(1-x**2)
        A=A+h*y
        x=x+h
    
    return 4*A

aire_du_disque()
```
